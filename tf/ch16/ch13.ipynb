```python
import tensorflow as tf
```

```python
dataset = tf.data.Dataset.from_tensor_slices(tf.range(10))
```

```python
for i in dataset.batch(7):
    print(i)
```

```python
for i in dataset.repeat(2):
    print(i)
```

```python
for i in dataset.repeat(3).batch(7,drop_remainder=True):
    print(i)
```

```python
dataset_batch = dataset.batch(3, drop_remainder=True)
```

```python
for i in dataset_batch:
    print(i)
```

```python
for i in dataset_batch.map(lambda x:x*2):
    print(i)
```

```python
for i in dataset_batch.map(lambda x:x*2).shuffle(buffer_size=4, seed=42):
    print(i)
```
